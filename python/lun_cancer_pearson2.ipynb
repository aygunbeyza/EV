```python
import pandas as pd
from scipy import stats
import numpy as np

x = votes_df_luca

# Take the remaining columns as values on the y-axis
y = luca_new

#  combine x and y
data = np.column_stack((x, y))

# Calculate Pearson correlation
correlation_matrix = np.corrcoef(data, rowvar=False)

# Get correlations between first column and other columns
correlations_with_x = correlation_matrix[0, 1:]

print(correlations_with_x)
```

```python
nan_count = np.sum(np.isnan(correlations_with_x))
non_nan_count = len(correlations_with_x) - nan_count
print("NaN olan sütun sayısı:", nan_count)
print("NaN olmayan sütun sayısı:", non_nan_count)
```

```python
# Sort correlations from largest to smallest
sorted_indices = np.argsort(correlations_with_x)[::-1]

# Get top 10 correlations and gene names
top_100_indices = sorted_indices[:100]


for idx in top_100_indices:
    gene = luca_new.columns[idx]  # +1 because the first column is on the x-axis
    correlation = correlations_with_x[idx]
    print(f"Gen: {gene}, Korelasyon Değeri: {correlation}")
```

```python
# Her bir korelasyon değeri için gen isimlerini bul
y_columns = luca_new.columns

for i, corr in enumerate(correlations_with_x):
    y_column_name = y_columns[i]
    print(f"Korelasyon değeri '{corr}' için ilgili gen: {y_column_name}")
```

```python
# NaN değerlere sahip olan korelasyonların indekslerini bul
nan_indices = np.where(np.isnan(correlations_with_x))[0]

# NaN değerlere sahip olan gen isimlerini bul
nan_gen_names = luca_new.columns[nan_indices]

print("Korelasyon değeri NaN olan genlerin isimleri:", nan_gen_names)
```

```python
# NaN değerlere sahip olan gen isimlerini virgülle ayrılmış şekilde ve tırnak içinde yazdır
nan_gen_names_str = ', '.join([f'"{gen}"' for gen in nan_gen_names])

print("Korelasyon değeri NaN olan genlerin isimleri:", nan_gen_names_str)
```

```python
# NaN değerlere sahip olan korelasyonları filtrele
valid_correlations = correlations_with_x[~np.isnan(correlations_with_x)]

# En yüksek korelasyon değerlerine sahip genlerin indekslerini al
top_100_indices = np.argsort(valid_correlations)[::-1][:100]

# En yüksek korelasyon değerlerine sahip genlerin adlarını ve korelasyon değerlerini al
top_100_genes = luca_new.columns[top_100_indices]
top_100_correlations = valid_correlations[top_100_indices]

# Genlerin adları ve korelasyon değerlerini içeren bir sözlük oluştur
top_100_genes_with_correlations = [{'Gen': gene, 'Korelasyon Değeri': corr} for gene, corr in zip(top_100_genes, top_100_correlations)]

print("En yüksek 100 genin adları ve korelasyon değerleri:")
for item in top_100_genes_with_correlations:
    print(item)
```

```python
# NaN değerleri hesaba katmadan en küçük ve en büyük korelasyon değerlerini bul
min_correlation = np.nanmin(correlations_with_x)
max_correlation = np.nanmax(correlations_with_x)

print("En küçük korelasyon değeri:", min_correlation)
print("En büyük korelasyon değeri:", max_correlation)
```
