```python
from scipy.io import mmread
import pandas as pd

lamp5s = mmread('/Users/beyzaaygun/Desktop/Bioinformatic/EV/2023-2024/python/lamp5s.mtx')


# Convert to DataFrame
df_lamp5 = pd.DataFrame(lamp5s.toarray())

# Read TSV files containing row and column names
row_names_lamp5 = pd.read_csv('cell_lamp.tsv', delimiter='\t')
column_names_lamp5 = pd.read_csv('gene.tsv', delimiter='\t')

# Add row and column names to the DataFrame
df_lamp5.index = row_names_lamp5.iloc[:, 0]
df_lamp5.columns = column_names_lamp5.iloc[:, 0]

df_lamp5.head()
```

```python
mac = mmread('/Users/beyzaaygun/Desktop/Bioinformatic/EV/2023-2024/python/macs.mtx')

df_mac = pd.DataFrame(mac.toarray())

row_names_mac = pd.read_csv('cell_mac.tsv', delimiter='\t')
column_names_mac = pd.read_csv('gene.tsv', delimiter='\t')

df_mac.index = row_names_mac.iloc[:, 0]
df_mac.columns = column_names_mac.iloc[:, 0]

df_mac.head()
```

```python
frames = [df_mac, df_lamp5]

concat_df = pd.concat(frames)

concat_df.head()
```

```python
# Finding the counts of "lamp5" and "macrophage" rows
lamp5_count = concat_df.index.value_counts().get('lamp5', 0)
macrophage_count = concat_df.index.value_counts().get('mac', 0)


print("Number of rows for lamp5:", lamp5_count)
print("Number of rows for macrophage:", macrophage_count)
```

```python
concat_df['y'] = ''

# While creating the 'y' column, set the value as 'positive' for rows with 'macrophage' and 'negative' for rows with 'lamp5'
concat_df.loc[concat_df.index == 'mac', 'y'] = 'positive'
concat_df.loc[concat_df.index == 'lamp5', 'y'] = 'negative'

concat_df.to_csv('concat_df.csv', index=True)
print(concat_df)
```
